```python

import matplotlib.pyplot as plt

student1 = {'chinese':78, 
            'english':80,
            'math':92,
            'history':75,
            'discover':85}

student2 = {'chinese':69, 
            'english':92,
            'math':74,
            'history':69,
            'discover':93}

subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())


figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,score1,'ro--',label='Sudent1')
axes.plot(subjects,score2,'bo-',label='Student2')


axes.set_title('Table Of Students')
axes.set_xlabel('Subject')
axes.set_ylabel('Scores')
axes.grid(axis='y')
axes.legend()

x_location = axes.get_xticks()

for i in x_location:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+1))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+1))

plt.show()
```
