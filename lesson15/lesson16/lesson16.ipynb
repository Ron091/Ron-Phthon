```python
import matplotlib.pyplot as plt
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors =['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,
        colors=colors,
        explode=[0.1,0,0,0],
        labels=labels,
        autopct='%1.1f%%',
        shadow=True,
        startangle=180) 

plt.show()
```
