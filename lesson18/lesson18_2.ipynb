```python
import numpy as np
import pandas as pd
import random
with open('names.txt',encoding='utf-8') as file:
    content = file.read()
names =content.split()
studentsName = random.choices(names,k=50)

score = np.random.randint(50,101,(50,5))
students_df = pd.DataFrame(score,
             columns=['國文','英文','數學','社會','自然'],
             index=studentsName)
```

```python
from pandas import DataFrame,Series
def fun3(s:Series):
    sum = s.sum()
    mean = s.mean()
    count = s[s<60].count()
    values = s[s<60].index.values
    return Series([sum,mean,count,values],index=['總分''平均','不及格數','不及格科目'])


students_df.apply(fun3,axis=1)
```
