```python
import tools
from tools import get_path,in_out_imformation,get_station_imfo
import pandas as pd
if __name__ == "__main__":
    name_list = get_path()
    stations_imfo = get_station_imfo()
    all_in_out = in_out_imformation(name_list,stations_imfo)
    station_in_out = pd.concat(all_in_out)
    station_in_out
```

```python
station_in_out1= station_in_out.sort_values(by="日期")
```

```python
station_in_out2 = station_in_out1.groupby("站名")[['入站','出站']]
```

```python
df2=station_in_out2.aggregate([("最多人數","max"),('最少人數',"min"),('中間值','median')])
df2
```

```python
serial=df2.stack(level=[0,1])
serial
```

```python
serial.index.names= ['站名','統計','出入站']
serial.unstack(level='站名')
```
