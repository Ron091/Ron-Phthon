```python
import tools
from tools import get_path,in_out_imformation,get_station_imfo
import pandas as pd
if __name__ == "__main__":
    name_list = get_path()
    stations_imfo = get_station_imfo()
    all_in_out = in_out_imformation(name_list,stations_imfo)
    station_in_out = pd.concat(all_in_out)
    print(station_in_out)
```

```python
years =station_in_out['日期'].dt.year
station_in_out.groupby(by=years)[['入站','出站']].mean()
```

```python
years_month =station_in_out['日期'].dt.strftime("%Y-%m")
station_in_out.groupby(by=years_month)[['入站','出站']].mean()
```
