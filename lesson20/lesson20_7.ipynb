```python
import tools
from tools import get_path,in_out_imformation,get_station_imfo
import pandas as pd
if __name__ == "__main__":
    name_list = get_path()
    stations_imfo = get_station_imfo()
    all_in_out = in_out_imformation(name_list,stations_imfo)
    station_in_out = pd.concat(all_in_out)
    station_in_out1= station_in_out.sort_values(by="日期")
station_in_out1
```

```python
def top(df, n=5, column='入站'):
    return df.sort_values(by=column,ascending=False)[:n]

top(station_in_out1)
```

```python
years_months = station_in_out1['日期'].dt.strftime('%Y-%m')
station_in_out2 = station_in_out1.groupby(by=[years_months,'站名'])[['入站','出站']].apply(top)
station_in_out2
```

```python

station_in_out2 = station_in_out1.groupby(by=[years_months,'站名'])[['入站','出站']].apply(top,n=1)
station_in_out2
```

```python
station_in_out2 = station_in_out1.groupby(by=[years_months,'站名'])[['入站','出站']].apply(top,n=1,column='出站')
station_in_out2
```
